# Введение
В качестве своего направления исследования я выбрал изучение статистики и имеющейся информации об онкологических заболеваниях. Тематика моего исследования очень актуальна, а также является очень личной темой для многих: почти все знают или потеряли кого-то дорогого, кто столкнулся с этим недугом. 

Каждая шестая смерть в мире вызвана ​​раком, что делает его второй ведущей причиной смерти (уступая только сердечно-сосудистым заболеваниям). По оценкам, только в 2016 году умерли 8.9 миллионов человек от различных форм рака. Институт по измерению показателей здоровья и оценке состояния здоровья (IHME) установил относительно небольшие погрешности по всему этому глобальному показателю: нижняя и верхняя оценки составляют от 8,75 до 9,1 млн. человек.

Прогресс в борьбе со многими другими причинами смерти людей и демографическими факторами роста численности населения, а также ожидаемой продолжительности жизни и старения населения означает, что общее число смертей от рака продолжает увеличиваться.

Национальный институт рака (NCI) определяет рак как совокупность заболеваний, при которых аномальные клетки могут делиться и распространяться в соседнюю ткань. Исходя из этого определения, рак может возникать во многих частях тела (приводя к различным видам рака, как будет показано ниже), а в некоторых случаях распространяется на другие части тела через кровь и лимфатические системы.



Что вызывает рак?
Так что же все таки вызывает рак? Рак появляется в результате превращения нормальных клеток в опухолевые в ходе многоэтапного процесса. Эти изменения происходят в результате воздействия генетических факторов человека и тремя категориями внешних факторов, включающих:
физические канцерогены (ультрафиолетовое и ионизирующее излучение)
химические канцерогены (асбест, компоненты табачного дыма, афлатоксины, и мышьяк (загрязнитель питьевой воды) )
биологические канцерогены (инфекции, вызываемые некоторыми вирусами, бактериями или паразитами)
Еще одним основополагающим фактором развития рака является возраст. С течением жизни человека заболеваемость раком резко возрастает, вероятнее всего, в связи с накоплением факторов риска развития определенных раковых заболеваний. Общее накопление рисков усугубляется тенденцией к снижению эффективности механизмов обновления клеток по мере старения человека.

## Что вызывает рак?
Рак возникает в результате преобразования нормальных клеток в опухолевые клетки в ходе многоэтапного процесса, в ходе которого предраковое поражение переходит в злокачественную опухоль. Эти изменения происходят в результате взаимодействия между генетическими факторами человека и тремя категориями внешних факторов, включающих:
- физические канцерогены, такие как ультрафиолетовое и ионизирующее излучение;
- химические канцерогены, такие как асбест, компоненты табачного дыма, афлатоксины (загрязнители пищевых продуктов) и мышьяк (загрязнитель питьевой воды);
- биологические канцерогены, такие как инфекции, вызываемые некоторыми вирусами, бактериями или паразитами.
ВОЗ через свое Международное агентство по изучению рака (МАИР) осуществляет классификацию канцерогенных факторов.
Еще одним основополагающим фактором развития рака является старение. С возрастом заболеваемость раком резко возрастает, вероятнее всего, в связи с накоплением факторов риска развития определенных раковых заболеваний. Общее накопление рисков усугубляется тенденцией к снижению эффективности механизмов обновления клеток по мере старения человека.
### Какие существуют факторы риска развития раковых заболеваний?
Основными факторами риска развития рака в мире являются употребление табака, употребление алкоголя, нездоровое питание и отсутствие физической активности; также факторами риска развития рака являются некоторые хронические инфекции, особенно в странах с низким и средним уровнем доходов. Канцерогенные инфекции, в том числе Helicobacter pylori, вирус папилломы человека (ВПЧ), вирус гепатита B, вирус гепатита C и вирус Эпштейна-Барра были названы причиной примерно 15% случаев рака, диагностированных в 2012 г.3.
Вирус гепатита B и C и некоторые типы ВПЧ повышают риск рака печени и шейки матки, соответственно. ВИЧ-инфекция значительно повышает риск развития рака, например рака шейки матки.


## Повышаются ли показатели смертности от рака?
Общая глобальная тенденция показывает,что несмотря на все большее число смертей от рака, индивидуальные показатели смертности снижаются. В 1990 году 161 человек из каждых 100 000 человек в мире умер от рака - к 2016 году этот показатель снизился до 134 на 100 000 человек.

В приведенной ниже таблице мы видим индивидуальную стандартизованную смертность среди всех типов рака. Смертность в данном случае измеряется как количество смертей на 100 000 человек.

Как и следовало ожидать: показатели трахеи, бронха и рака легких наиболее высоки в мире на уровне 26 на 100 000 человек. Но все же показатели смертности снизились: примерно с 29 на 100 000 в 1990 году, причем в некоторых странах наблюдается еще больший спад (в США ставки упали с 53 до 41 на 100 000).

В некоторых случаях мы наблюдаем резкое снижение за последние несколько десятилетий. Наиболее заметным является рак желудка, смертность которого снизилась более чем на 40 процентов с 22 до 13 на 100 000 человек. С 1990 года показатели рака пищевода и шейки матки снижались примерно на треть


```python
import numpy as np
```

```python
df = pd.read_csv('cancer-death-rates-by-type.csv')
df.head()
```

```python
data = []
n = df.columns.size - 3
for i in range(n):
    data.append([])
```

```python
years = df['Year'].unique()
for year in range(years.min(), years.max() + 1):
    i = 0
    for col in df.loc[:, 'Prostate cancer (per 100,000)'::]:
        data[i % n].append(df[df['Year'] == year][col].mean())
        i += 1
```

```python
columns = df.loc[:, 'Prostate cancer (per 100,000)'::].columns
data.sort(key=lambda x: np.mean(x), reverse=True)
```

```python
plt.figure(figsize=(16, 9))
figures = []
for i in range(n // 3 ):
    figures.append(plt.plot(years, data[i], linestyle='-.' , linewidth=3, label=columns[i]))
plt.legend()
plt.ylabel("Коэффициент")
plt.title("Коэффициенты смертности от рака по типу (на 100 000), в мире.\n", fontsize=25)
plt.xlabel("Год")
plt.grid()

plt.show()
```

## Выживаемость
Несмотря на увеличение смертности от рака в мире, смертность от рака во многих видах онкологических заболеваний снизилась. Это отчасти объясняется более ранним выявлением и улучшенным лечением: оба этих фактора способствовали значительному увеличению показателей выживаемости рака во всем мире.

Опираясь на имеющиеся исследования изменений в показателях выживаемости на пять лет описанные в ежегодных отчетах ВОЗ я заметил, что показатели выживаемости от всех видов рака увеличились с 50 до 67 процентов. Это связано с комбинацией как раннего выявления, так и улучшения лечения.

Однако это изменение значительно различается по типу рака. Прогресс в некоторых случаях был очень большим: показатели выживаемости от рака предстательной железы, например, увеличились с 68 до почти 99 процентов за период с 1970-х по 2007-2013. На другом конце шкалы пятилетние показатели выживаемости маточных и шейных маточных раков - они фактически снизились с 1970-х годов в обратную сторону (с 86.9 % до 82 %). Мы по-прежнему наблюдаем значительные различия в показателях выживаемости в целом: у некоторых, таких как простата и щитовидная железа, уровень выживаемости составляет 98-99 процентов; а те виды рака, что находились ранее в нижней части спектра (поджелудочная железа, печень, легкие и пищевод) имеют вероятность менее 20%.


Национальный институт рака США также опубликовал данные о пятилетних показателях выживаемости. Эти данные доступны в диаграмме ниже. В целом мы наблюдаем сходные тенденции: у предстательной железы, щитовидной железы, кожи и рака молочной железы самые высокие показатели выживаемости, а поджелудочная железа, печень, легкие и пищевод - самые низкие.

В течении подготовки проекта я научился проводить собственные исследования и доносить их результаты до других, визуализировать данные, разрабатывать интерактивные веб-интерфейсы, собирать и структурировать любые данные, которые есть в интернете. Таким образом этот проект немного пересекается с работой data scientist'a: от постановки задачи, исследований и предобработкой данных, поиск закономерностей в данных, проверка гипотез, визуализация данных и построение каких-то выводов.
## Источники данных и используемая литература

Institute of Health Metrics and Evaluation (IHME), Global Burden of Disease (GBD). Какие данные были использованы: показатели смертности и индекс DALY по всем факторам риска и причинам. Географический охват - глобальный, во всех регионах и странах. Период времени: большинство показателей, доступных с 1990 года. Доступно по адресу - http://ghdx.healthdata.org/gbd-results-tool
World Health Organization (WHO) Global Health Observatory (GHO). Какие данные были использованы: Причины смертности по возрасту и полу. Географический охват - глобальный, во всех регионах и странах. Промежуток времени: большинство показателей, доступных с 2000 по 2015 год в 5-летних частях. Доступно по адресу - http://www.who.int/healthinfo/global_burden_disease/

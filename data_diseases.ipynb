### Причины смерти существенно различаются в зависимости от страны и уровня доходов во всем мире. Эти случаи смерти за 2016 год показаны на диаграмме ниже, которые представляют собой процентную долю от общего числа смертей. НИЗ(неинфекционные заболевания) не только доминируют в показателях смертности на глобальном уровне, но и являются причиной большинства смертей в странах с высоким уровнем дохода. На втором месте идут раковые заболевания. О раке известно, что это не одна болезнь, а, по крайней мере, 200 - причем каждая из них имеет свои симптомы, методы диагностики и лечения.
```python
import pandas as pd
import numpy as np
import seaborn as sns
from pylab import rcParams
rcParams['figure.figsize'] = 8, 5

from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('./deathes.csv')
```

```python
df.head()
```

```python
df[pd.isnull(df['Code']) == True]
```

```python
df['Year'].value_counts()
```

```python
df2016 = df[df['Year'] == 2016]
```

```python
df_new = df2016.loc[:,'Dementia':'Terrorism (deaths)':]
```

```python
df_new.insert(loc=0, column='Country', value=df2016['Entity'])
```

```python
df_new = df_new.drop(['Execution (deaths)'], axis=1)
```

```python
df_new.head()
```

```python
res = df_new.apply(lambda x: x.fillna(0.0))
```

```python
res.head()
```

```python
sum_val = (res.loc[:, 'Dementia'::].sum().astype('int').sum()).astype('int')
```

```python
plot = (res.loc[:, 'Dementia'::].sum()*100.0/sum_val).sort_values(ascending=False).plot(kind='bar', figsize=(15,5), fontsize=20)
plot.set_xlabel("Diseases", fontsize=30)
plot.set_ylabel("Percentage", fontsize=30)

plt.show()
```

### Рак является одной из основных причин смерти в мире; так, в 2016 г. от этого заболевания умерли 8,9 млн человек. Рак становится причиной практически каждой шестой смерти в мире

### Рак определен Национальным Институтом рака (NCI) как собрание заболеваний, в которых анормалные клетки могут делиться и распространяться к близрасположенной ткани. # Факторы риска 

## Cравнение факторов риска смерти

### Она включает факторы риска по четырем широким категориям: поведенческие, экологические, профессиональные и метаболические риски.

### В приведенной ниже диаграмме мы видим количество смертей, связанных с конкретными факторами риска в 2016 году. Эти данные измеряются по всем возрастным группам и обоим полам. Здесь мы видим, что есть несколько доминирующих факторов риска для смерти: в частности, те, которые относятся к питанию и активности, факторы образа жизни (в том числе артериального давления, физической активности, индекса массы тела, сахара в крови, и поступление с пищей); курение; загрязнение воздуха (наружного и внутреннего); факторов окружающей среды в том числе питьевой воде и санитарии; и безопасный секс (по профилактике ВИЧ/СПИДа).

### Вклад конкретных факторов риска существенно варьируется в зависимости от страны. Для большинства стран с высоким уровнем дохода доминирующими факторами риска являются факторы, связанные со здоровым питанием, курением и употреблением алкоголя. Другие факторы риска, такие как чистая вода, санитария и истощение детей или задержка роста, очень низки. В странах с низким уровнем дохода верно обратное.

### Рак возникает в результате преобразования нормальных клеток в опухолевые клетки в ходе многоэтапного процесса, в ходе которого предраковое поражение переходит в злокачественную опухоль. Эти изменения происходят в результате взаимодействия между генетическими факторами человека и тремя наиболее существенными категориями внешних факторов, включающих:
### * физические канцерогены, такие как ультрафиолетовое и ионизирующее излучение;

### * химические канцерогены, такие как асбест, компоненты табачного дыма, афлатоксины (загрязнители пищевых продуктов) и мышьяк (загрязнитель питьевой воды);

### *биологические канцерогены, такие как инфекции, вызываемые некоторыми вирусами, бактериями или паразитами.



## Смертность от рака в разбивке по возрасту

### Еще одним основополагающим фактором развития рака является старение. С возрастом заболеваемость раком резко возрастает, вероятнее всего, в связи с накоплением факторов риска развития определенных раковых заболеваний. Общее накопление рисков усугубляется тенденцией к снижению эффективности механизмов обновления клеток по мере старения человека. В приведенной ниже визуализации мы видим, как показатели смертности от рака варьируются в зависимости от возрастной группы. Как мы видим, самые высокие показатели смертности отмечаются среди лиц в возрасте старше 70 лет и составляют почти 1000 смертей на 100 000 человек. Ставки в категории 50-69 лет более чем в три раза ниже и составляют чуть менее 300 на 100 000 человек. Для детей в возрасте до 14 лет, цены на мировом уровне-менее 5 на 100 000.
